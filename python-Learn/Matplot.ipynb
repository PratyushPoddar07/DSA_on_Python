## MatplotLib Tutorial
 - MatplotLib is a plotting library for the python programming language and ites numerical mathematics extension NumPy. It Provides an objects-oriented API for embedding plots into application using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

 - Some of the major pros of Matplotlib are:
   - Generally easy to get started for  simple plotys
   - Support for custom labels and texts
   - great control of every element in a figure
   - High - quality output in many formats
   - Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y = x*x
```

```python
## plt plot
plt.plot(x,y,'b*--')
```

```python
## Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")

plt.plot(x,y)
```

```python
# Subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5 * np.pi ,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


# Set up a subplot grid that has height 2 and weight 1,
# and set the first such subplot as active
plt.subplot(2,2,1)
plt.plot(x,y_sin,'b--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'d--')
plt.title('Cosine')
```

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align ='center')
plt.bar(x2,y2,color = 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

```

## Histogram


```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
```

## Box Plot using Matplotlib


```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert = True,patch_artist = True);
```

```python
# Data to plot
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

# plot
plt.pie(sizes,explode = explode ,labels = labels,colors = colors,autopct ='%1.1f%',shadow =True)
plt.axis('equal')
```
